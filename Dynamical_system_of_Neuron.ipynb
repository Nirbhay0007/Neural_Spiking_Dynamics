### Derivative of \( Q = C V \)

If both \( C \) and \( V \) depend on time \( t \):

$$
\frac{dQ}{dt} = \frac{d}{dt}(C V) = C \frac{dV}{dt} + V \frac{dC}{dt}
$$

If \( C \) is constant (does not change with time):

$$
\frac{dQ}{dt} = C \frac{dV}{dt}
$$

$$   I_{ion} = \underbrace{g_{ion}(E_{ion} - V)}_{\text{Driving force}}$$

### In Real world this conductance is not constant , which means many of the channels opening  and closing  depend on the membrane voltage it self so we have to come up with a expression where `g` is the function of voltage compeleting our discription

    * Fraction of open ion channel is same thing as [ the probabilty of channel being open ]

\end{cases}
$$

# üß† Reduction of the Gating Variable Equation

In the Hodgkin‚ÄìHuxley model, each gating variable (like \( n, m, h \)) follows the dynamic:

$$
\frac{dn}{dt} = \alpha_n(V)(1 - n) - \beta_n(V)n
$$

This describes two competing processes:
$$ \alpha_n(V)(1 - n) : \text{opening of gates } $$
$$  \beta_n(V)n : \text{closing of gates}  $$

---

## ‚öôÔ∏è Step 1: Expand and Rearrange
$$
\frac{dn}{dt} = \alpha_n(V) - (\alpha_n(V) + \beta_n(V))n
$$

Now we can see \( n \) decreases at a rate proportional to itself and increases toward a voltage-dependent constant.

---

## ‚öôÔ∏è Step 2: Factor Common Terms
$$
\frac{dn}{dt} = (\alpha_n(V) + \beta_n(V)) \left[ \frac{\alpha_n(V)}{\alpha_n(V) + \beta_n(V)} - n \right]
$$

This shows that the rate of change of \( n \) depends on how far it is from a certain **target value**.

---

## ‚öôÔ∏è Step 3: Define Steady-State and Time Constant
Let:
$$
n_\infty(V) = \frac{\alpha_n(V)}{\alpha_n(V) + \beta_n(V)}, 
\quad
\tau_n(V) = \frac{1}{\alpha_n(V) + \beta_n(V)}
$$

Substitute them into the equation:

$$
\frac{dn}{dt} = \frac{n_\infty(V) - n}{\tau_n(V)}
$$

---

## üåä Interpretation
- \( n_\infty(V) \): steady-state (equilibrium) value toward which \( n \) relaxes.  
- \( \tau_n(V) \): time constant controlling how fast it relaxes.  

Thus the solution takes the familiar exponential form:
$$
n(t) = n_\infty(V) + (n_0 - n_\infty(V)) e^{-t / \tau_n(V)}
$$

## üß† The Setup

In the **Hodgkin‚ÄìHuxley model**, we have **three gating variables**:

$$
n(t), \quad m(t), \quad h(t)
$$

Each follows the same type of equation:

$$
\frac{dx}{dt} = \frac{x_\infty(V) - x}{\tau_x(V)}
$$

where \( x \) is \( n \), \( m \), or \( h \).

* x_infty(V) : steady-state (final) value at a given voltage  
*  tau_x(V) : time constant ‚Äî how quickly \( x \) reaches equilibrium

---



## ‚öôÔ∏è What the Graph Shows

Each curve shows **how fast** each gating variable responds when voltage changes:

| Variable | œÑ (time constant) | Meaning |
|-----------|-------------------|----------|
| tau_n = 3.7 { ms}  | **Slow** | Potassium gate opens slowly |
| tau_m = 0.009 \text{ ms}  | **Extremely fast** | Sodium activation gate reacts almost instantly |
| tau_h = 0.71 \text{ ms}  | **Medium** | Sodium inactivation gate reacts moderately fast |

---

## üöÄ The Key Idea ‚Äî Why \( m(t) \) and \( h(t) \) Can Be **Simplified**

Look at the **middle graph** ‚Äî \( m(t) \) rises to its final value **almost instantly** because:

$$
\tau_m = 0.009 \text{ ms}
$$

is extremely small.

That means:

$$
m(t) \approx m_\infty(V)
$$

In plain English:

> The sodium activation variable \( m \) changes so fast that it‚Äôs practically *always* at its steady-state value for the current voltage.

---

### ‚ö° So What Happens in the Equations?

Normally, we‚Äôd have to **solve**:

$$
\frac{dm}{dt} = \frac{m_\infty(V) - m}{\tau_m(V)}
$$

But since  tau_m  is very small,  m  catches up to  m_infty(V)  almost instantly.

So we **approximate**:

$$
m(t) \approx m_\infty(V)
$$

and **remove** the differential equation for \( m \).

This greatly simplifies the system ‚Äî instead of tracking how \( m \) *changes*,  
we just assume it‚Äôs already at its equilibrium for that voltage.

---

### üß© What About \( h(t) \) and \( n(t) \)?

*  h(t) : has a moderate  tau_h  ‚Üí we still model it dynamically.  
* n(t) : slowest tau_n = 3.7{ ms}  ‚Üí we definitely need its time evolution.

So:

* Keep **n** and **h** as dynamic variables.  
* Approximate **m ‚âà m‚àû(V)**.

---

## ‚ú≥Ô∏è Result ‚Äî The Simplified Equations

After this approximation, the **membrane current equations** simplify:

$$
I_{\text{Na}} = g_{\text{Na}} m^3 h (V - E_{\text{Na}})
\quad \Rightarrow \quad
I_{\text{Na}} \approx g_{\text{Na}} \, m_\infty(V)^3 \, h \, (V - E_{\text{Na}})
$$

Now only \( h \) and \( n \) need differential equations ‚Äî  
\( m \) is replaced by \( m_\infty(V) \), a direct function of voltage.

---

### üß† In Short

| Variable | Time Constant | Behavior | What We Do |
|-----------|----------------|-------------|----------------|
| \( m \) | Very small | Instantaneous | Replace \( m(t) \) with \( m_\infty(V) \) |
| \( h \) | Medium | Dynamic | Keep differential equation |
| \( n \) | Slow | Dynamic | Keep differential equation |




### üß† Why $h$ Can Be Removed

In many *reduced* Hodgkin‚ÄìHuxley models we **remove $m$** and **also remove $h$** (or replace $h(t)$ by its steady-state value) because of these simplifying assumptions:

* **$m$** is extremely fast ‚Üí set $m(t) \approx m_\infty(V)$
* **$h$** is either fast enough (or tightly voltage-dependent) that we approximate it by its steady-state value $h_\infty(V)$ (or by a simple function of $V$ or $n$)
    ‚Üí this eliminates the differential equation for $h$ and replaces $h(t)$ by a direct function of $V$
* The result: only **membrane voltage $V(t)$** and one gating variable (typically $n(t)$) remain dynamic ‚Äî giving a **2D system** that‚Äôs much simpler to analyze and still captures main excitability behavior.

> ‚öñÔ∏è *Tradeoff:* you lose some fine timescale structure of sodium inactivation, but still capture the qualitative spike dynamics.

---

### ‚öôÔ∏è Final Two Differential Equations
*(Standard 2-variable Hodgkin‚ÄìHuxley reduction)*

Assume $m \approx m_\infty(V)$ and $h \approx h_\infty(V)$.
Then the reduced system becomes:

$$
\boxed{
C_m \frac{dV}{dt}
= - g_{Na} \, m_\infty(V)^3 \, h_\infty(V) \, (V - E_{Na})
- g_K \, n^4 (V - E_K)
- g_L (V - E_L)
+ I_{\text{ext}}
}
$$

$$
\boxed{
\frac{dn}{dt} = \frac{n_\infty(V) - n}{\tau_n(V)}
}
$$

---

### üß© Parameter Definitions

* $C_m$: membrane capacitance
* $g_{Na}, g_K, g_L$: maximum conductances
* $E_{Na}, E_K, E_L$: reversal potentials

And the voltage-dependent steady-state functions:

$$
m_\infty(V) = \frac{\alpha_m(V)}{\alpha_m(V) + \beta_m(V)}, \quad
h_\infty(V) = \frac{\alpha_h(V)}{\alpha_h(V) + \beta_h(V)}, \quad
n_\infty(V) = \frac{\alpha_n(V)}{\alpha_n(V) + \beta_n(V)}
$$

Time constant for $n$:

$$
\tau_n(V) = \frac{1}{\alpha_n(V) + \beta_n(V)}
$$

---

‚úÖ **Result:**
We now have a **two-dimensional system** of ODEs ‚Äî
$V(t)$ and $n(t)$ are dynamic,
while $m$ and $h$ are algebraic functions of $V$.
